## Differential Privacy Equations

## 1)
$$
\Delta S = \frac{2C}{|D_i|}
$$

where:

$$ |D_i| = 15 \quad \text{(the number of samples per each client is same )} $$


$$ C \quad \text{(the clipping parameter)} $$

to reduce the influence of a single model on the global model




## 2)

To ensure $(\epsilon, \delta)$-DP for each client in the uplink during one exposure, we use the following equation to calculate the noise standard deviation ($\sigma$):

$$
\sigma = \frac{c \cdot \Delta S}{\epsilon}
$$

where:

*  $\sigma$ represents the noise standard deviation. [ NOISE MULTIPLIER ]
*  $c$ is a constant calculated using the formula below.
*  $\Delta S$  is the sensitivity of the query.
*  $\epsilon$ is the privacy parameter.

The constant $c$ is determined using the following equation:

$$
c = \sqrt{2 \ln\left(\frac{1.25}{\delta}\right)}
$$

where:

*  $\delta$ is another privacy parameter representing the probability of a privacy breach.

```python
import numpy as np
import pandas as pd
import tensorflow_privacy as tfp
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import math

# Create 4 clients
num_clients = 4
rows_per_client = 30
iterations = X_train.shape[0] // (num_clients * rows_per_client)

# Differential privacy parameters
def get_noise_multiplier(epsilon, delta, l2_norm_clip, rows_per_client):
    return (math.sqrt(2 * math.log(1.25 / delta)) * 2 * l2_norm_clip) / (epsilon * rows_per_client)

epsilon = 1000000000  # Privacy loss parameter
l2_norm_clip = 1.0      # Gradient clipping threshold
learning_rate = 0.015
epochs = 32
delta = 1e-5          # Delta (probability of privacy violation)
total_steps = iterations * epochs  # Total number of training steps

# Function to fit logistic regression model with differential privacy
def fit_logistic_regression(X, y, noise_multiplier):
    model = LogisticRegression(solver='lbfgs', max_iter=epochs)
    model.fit(X, y)

    # Add Gaussian noise to model parameters for differential privacy
    for i in range(len(model.coef_[0])):
        noise = np.random.normal(0, noise_multiplier, size=model.coef_[0].shape)
        model.coef_[0] += noise

    return model.coef_, len(y)  # Return model coefficients and sample size

# Federated learning process
def client_update(X, y, noise_multiplier):
    return fit_logistic_regression(X, y, noise_multiplier)

# FedAvg Aggregation algorithm
def fedavg_aggregate(client_weights):
    total_samples = sum([num_samples for _, num_samples in client_weights])

    # Weighted average of the weights
    averaged_weights = [
        np.sum([client_weight[0] * (num_samples / total_samples)
                for client_weight, num_samples in client_weights], axis=0)
    ]
    return averaged_weights

# Initialize global model weights
global_weights = np.zeros((X_train.shape[1],))  # Initialize with zeros

for iteration in range(iterations):
    print(f"Iteration {iteration + 1}/{iterations}")
    client_weights = []

    # Update models on each client
    for client in range(num_clients):
        print(f"Client {client + 1}")
        start = 120 * iteration + client * rows_per_client
        end = start + rows_per_client
        client_data = (X_train[start:end], y_train[start:end])

        noise_multiplier = get_noise_multiplier(epsilon, delta, l2_norm_clip, rows_per_client)
        client_weight = client_update(*client_data, noise_multiplier)
        client_weights.append(client_weight)

    # Aggregate client weights using FedAvg
    averaged_weights = fedavg_aggregate(client_weights)
    global_weights = averaged_weights[0]  # Update global weights
    print(f"Iteration {iteration + 1}/{iterations} completed")

# Evaluate the final global model
# Since we only have coefficients, we need to use them to make predictions
y_pred = (X_test @ global_weights > 0.5).astype(int)  # Use test data for evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the global model: {accuracy * 100:.2f}%")

print(f"Epsilon for differential privacy: {epsilon}, Delta: {delta}")
```

```python
import numpy as np
import pandas as pd
import tensorflow_privacy as tfp
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import math
import tenseal as ts



# Set up parameters and contexts for CKKS encryption
def create_context():
    # CKKS encryption parameters
    # For TenSEAL v4.1.0 and above, 'scale' is no longer a direct argument.
    context = ts.context(ts.SCHEME_TYPE.CKKS, poly_modulus_degree=8192)
    # Set the scale using the global_scale method.
    context.global_scale = 2**40
    context.generate_galois_keys()
    return context

context = create_context()

# Create 4 clients
num_clients = 4
rows_per_client = 30
iterations = X_train.shape[0] // (num_clients * rows_per_client)

# Differential privacy parameters
def get_noise_multiplier(epsilon, delta, l2_norm_clip, rows_per_client):
    return (math.sqrt(2 * math.log(1.25 / delta)) * 2 * l2_norm_clip) / (epsilon * rows_per_client)

epsilon = 20  # Privacy loss parameter
l2_norm_clip = 1.0      # Gradient clipping threshold
learning_rate = 0.015
epochs = 100
delta = 1e-5          # Delta (probability of privacy violation)
total_steps = iterations * epochs  # Total number of training steps

# Function to fit logistic regression model with differential privacy
def fit_logistic_regression(X, y, noise_multiplier):
    model = LogisticRegression(solver='lbfgs', max_iter=epochs)
    model.fit(X, y)

    # Add Gaussian noise to model parameters for differential privacy
    for i in range(len(model.coef_[0])):
        noise = np.random.normal(0, noise_multiplier, size=model.coef_[0].shape)
        model.coef_[0] += noise

    return model.coef_, len(y)  # Return model coefficients and sample size

# Function to encrypt client weights using CKKS
def encrypt_weights(weights, context):
    encrypted_weights = ts.ckks_vector(context, weights)
    return encrypted_weights

# Function to decrypt weights
def decrypt_weights(encrypted_weights, context):
    decrypted_weights = encrypted_weights.decrypt()
    return decrypted_weights

# Federated learning process
def client_update(X, y, noise_multiplier, global_weights):
    # Train the model with the global weights
    model = LogisticRegression(solver='lbfgs', max_iter=epochs)
    model.coef_ = np.array([global_weights])  # Set initial coefficients to global weights
    model.fit(X, y)

    # Add Gaussian noise to model parameters for differential privacy
    for i in range(len(model.coef_[0])):
        noise = np.random.normal(0, noise_multiplier, size=model.coef_[0].shape)
        model.coef_[0] += noise

    # Encrypt the updated weights before returning
    updated_weights = encrypt_weights(model.coef_[0], context)
    return updated_weights, len(y)

# FedAvg Aggregation algorithm
def fedavg_aggregate(encrypted_client_weights):
    # Perform homomorphic addition of encrypted weights
    aggregated_weights = encrypted_client_weights[0][0]  # Start with the first client's weights

    for client_weight, _ in encrypted_client_weights[1:]:
        aggregated_weights += client_weight  # Homomorphically add the encrypted weights

    return aggregated_weights  # Return the aggregated encrypted weights

# Initialize global model weights
global_weights = np.zeros((X_train.shape[1],))  # Initialize with zeros

print(f" sigma : {get_noise_multiplier(epsilon, delta, l2_norm_clip, rows_per_client)}")

for iteration in range(iterations):
    print(f"Iteration {iteration + 1}/{iterations}")
    client_weights = []


    # Update models on each client
    for client in range(num_clients):
        print(f"Client {client + 1}")
        start = 120 * iteration + client * rows_per_client
        end = start + rows_per_client
        client_data = (X_train[start:end], y_train[start:end])

        noise_multiplier = get_noise_multiplier(epsilon, delta, l2_norm_clip, rows_per_client)
        client_weight = client_update(*client_data, noise_multiplier, global_weights)
        client_weights.append(client_weight)

    # Aggregate encrypted client weights using FedAvg
    aggregated_encrypted_weights = fedavg_aggregate(client_weights)

    # Decrypt the aggregated weights to update the global model
    global_weights = decrypt_weights(aggregated_encrypted_weights, context)

    print(f"Iteration {iteration + 1}/{iterations} completed")

# Evaluate the final global model
y_pred = (X_test @ global_weights > 0.5).astype(int)  # Use test data for evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the global model: {accuracy * 100:.2f}%")

print(f"Epsilon for differential privacy: {epsilon}, Delta: {delta}")
```

```python
import os
import numpy as np
import pandas as pd
import tensorflow_privacy as tfp
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import math
import tenseal as ts
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
from Crypto.Util.Padding import pad, unpad
import base64

# Step 1: Create keypair and save keys for CKKS (shared keys)
def ckks_create_keypair(key_directory):
    context = ts.context(ts.SCHEME_TYPE.CKKS, poly_modulus_degree=8192)
    context.global_scale = 2**40

    # Generate shared keys
    secret_key = context.secret_key()
    public_key = context.public_key()

    os.makedirs(key_directory, exist_ok=True)

    with open(os.path.join(key_directory, "shared_context.txt"), "wb") as f:
        f.write(context.serialize())

    # Serialize and save secret key
    with open(os.path.join(key_directory, "shared_private_key.txt"), "wb") as f:
        f.write(context.serialize(save_secret_key=True))

    return public_key, secret_key, context

# Step 2: Load Shared Public Key
def ckks_load_public_key(key_directory):
    with open(os.path.join(key_directory, "shared_context.txt"), "rb") as f:
        context_serialized = f.read()
        context = ts.context_from(context_serialized)
    return context  # Return the context containing the shared public key

# Step 3: Load Shared Private Key
def ckks_load_private_key(key_directory):
    with open(os.path.join(key_directory, "shared_private_key.txt"), "rb") as f:
        context_serialized = f.read()
        context = ts.context_from(context_serialized)
    return context  # Return the context containing the shared secret key


def ckks_decrypt_weights(encrypted_weights, private_key):
    encrypted_weights.link_context(private_key)
    decrypted_weights = encrypted_weights.decrypt()
    return decrypted_weights



# Step 4: Create and save AES keys for each client
def aes_create_keys(num_clients, key_directory):
    aes_keys = {}
    os.makedirs(key_directory, exist_ok=True)

    for client_id in range(1, num_clients + 1):
        key = get_random_bytes(16)  # Generate a random AES key (16 bytes for AES-128)
        aes_keys[f'client_{client_id}'] = key

        # Save each key in a file
        with open(os.path.join(key_directory, f'aes_key_client_{client_id}.bin'), 'wb') as f:
            f.write(key)

    return aes_keys

# AES Encryption Functions with key loading
def aes_load_key(client_id, key_directory):
    with open(os.path.join(key_directory, f'aes_key_client_{client_id}.bin'), 'rb') as f:
        key = f.read()
    return key

def aes_encrypt(data, key):
    cipher = AES.new(key, AES.MODE_CBC)
    ct_bytes = cipher.encrypt(pad(data, AES.block_size))
    iv = base64.b64encode(cipher.iv).decode('utf-8')
    ct = base64.b64encode(ct_bytes).decode('utf-8')
    return iv, ct

def aes_decrypt(iv, ct, key):
    iv = base64.b64decode(iv)
    ct = base64.b64decode(ct)
    cipher = AES.new(key, AES.MODE_CBC, iv)
    pt = unpad(cipher.decrypt(ct), AES.block_size)
    return pt

# Encrypting Weights Function with CKKS and AES
def ckks_aes_encrypt_weights(weights, public_key, aes_key):
    # Homomorphic encryption
    encrypted_weights = ts.ckks_vector(public_key, weights)

    # Convert weights to bytes for AES encryption
    weight_bytes = encrypted_weights.serialize()  # Get byte representation
    iv, ct = aes_encrypt(weight_bytes, aes_key)  # AES encryption
    return iv, ct, encrypted_weights


# Set up parameters and contexts for CKKS encryption
key_directory = "keys"
public_key, private_key, context = ckks_create_keypair(key_directory)

# Load the shared public key for client use
public_key = ckks_load_public_key(key_directory)

# Load the shared private key for decryption on the server side
private_key = ckks_load_private_key(key_directory)

# Create AES keys for each client
num_clients = 4
aes_keys = aes_create_keys(num_clients, key_directory)

# Create 4 clients (assuming X_train and y_train are defined)
rows_per_client = 30
iterations = X_train.shape[0] // (num_clients * rows_per_client)

# Differential privacy parameters
def get_noise_multiplier(epsilon, delta, l2_norm_clip, rows_per_client):
    return (math.sqrt(2 * math.log(1.25 / delta)) * 2 * l2_norm_clip) / (epsilon * rows_per_client)

epsilon = 20  # Privacy loss parameter
l2_norm_clip = 1.0      # Gradient clipping threshold
learning_rate = 0.015
epochs = 100
delta = 1e-5          # Delta (probability of privacy violation)
total_steps = iterations * epochs  # Total number of training steps

# Function to fit logistic regression model with differential privacy
def fit_logistic_regression(X, y, noise_multiplier):
    model = LogisticRegression(solver='lbfgs', max_iter=epochs)
    model.fit(X, y)

    # Add Gaussian noise to model parameters for differential privacy
    for i in range(len(model.coef_[0])):
        noise = np.random.normal(0, noise_multiplier, size=model.coef_[0].shape)
        model.coef_[0] += noise

    return model.coef_, len(y)  # Return model coefficients and sample size

# Federated learning process
def client_update(X, y, noise_multiplier, global_weights, client_id):
    # Load AES key for the client
    aes_key = aes_load_key(client_id, key_directory)

    # Train the model with the global weights
    model = LogisticRegression(solver='lbfgs', max_iter=epochs)
    model.coef_ = np.array([global_weights])  # Set initial coefficients to global weights
    model.fit(X, y)

    # Add Gaussian noise to model parameters for differential privacy
    for i in range(len(model.coef_[0])):
        noise = np.random.normal(0, noise_multiplier, size=model.coef_[0].shape)
        model.coef_[0] += noise

    # Encrypt the updated weights before returning
    iv, ct, _ = ckks_aes_encrypt_weights(model.coef_[0], public_key, aes_key)
    return iv, ct, len(y)


# FedAvg Aggregation algorithm
def fedavg_aggregate(encrypted_client_weights):
    total_samples = 0  # Initialize total samples
    aggregated_weights = None  # Initialize aggregated weights

    # First, iterate through all clients to calculate total samples
    for (iv, ct, sample_size) in encrypted_client_weights:
        total_samples += sample_size  # Accumulate total samples

    # Second, iterate again to compute the weighted average
    for client_id in range(1, len(encrypted_client_weights) + 1):
        (iv, ct, sample_size) = encrypted_client_weights[client_id - 1]  # Adjust for 0-indexing
        aes_key = aes_load_key(client_id, key_directory)

        # Decrypt using AES to get the encrypted CKKS vector bytes
        encrypted_ckks_bytes = aes_decrypt(iv, ct, aes_key)

        # Deserialize the encrypted CKKS vector bytes into a CKKSVector object
        decrypted_ckks_weight = ts.ckks_vector_from(context, encrypted_ckks_bytes)  # Use context for deserialization

        # Calculate the contribution to aggregated weights
        if aggregated_weights is None:
            aggregated_weights = decrypted_ckks_weight * (sample_size / total_samples)  # Initialize for the first client
        else:
            # Add the weighted contribution to aggregated_weights
            aggregated_weights += decrypted_ckks_weight * (sample_size / total_samples)

    return aggregated_weights  # Return the aggregated encrypted weights


# Initialize global model weights
global_weights = np.zeros((X_train.shape[1],))  # Initialize with zeros

for iteration in range(iterations):
    print(f"Iteration {iteration + 1}/{iterations}")
    client_weights_encrypted = []

    # Update models on each client
    for client in range(num_clients):
        print(f"Client {client + 1}")
        start = 120 * iteration + client * rows_per_client
        end = start + rows_per_client
        client_data = (X_train[start:end], y_train[start:end])

        noise_multiplier = get_noise_multiplier(epsilon, delta, l2_norm_clip, rows_per_client)
        client_weight = client_update(*client_data, noise_multiplier, global_weights, client + 1)
        client_weights_encrypted.append(client_weight)

    # Aggregate encrypted client weights using FedAvg
    aggregated_encrypted_weight = fedavg_aggregate(client_weights_encrypted)

    # Decrypt the aggregated weights to update the global model
    global_weights = ckks_decrypt_weights(aggregated_encrypted_weight, private_key)

    print(f"Iteration {iteration + 1}/{iterations} completed")

# Evaluate the final global model
y_pred = (X_test @ global_weights > 0.5).astype(int)  # Use test data for evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the global model: {accuracy * 100:.2f}%")

print(f"Epsilon for differential privacy: {epsilon}, Delta: {delta}")
```
